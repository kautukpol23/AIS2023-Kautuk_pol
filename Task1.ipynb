```python
dict1={'jan': 1200, 'feb': 1350, 'mar': 1375, 'apr': 1450, 'may': 1500, 'jun': 1250, 'jul': 75, 'aug': 125, 'sep': 110, 'oct': 100, 'nov': 85, 'dec': 1150}
dict1
```

```python
def Task1(capacity,demand):
    d=demand
    cap=int(capacity)
    frst=df['Cap'].loc[df.index[1]]
    df['Cap'] = df['Cap'].replace({frst:cap})
    
    key=d.keys()
    valu=list(d.values())
    for i in range(len(key)):
        df.loc[i, 'Dem'] = valu[i]
        if valu[i]>cap :
            df.loc[i, 'Extra_Qty'] = valu[i]-cap
            df.loc[i, 'Rem_Cap'] = 0
            
        elif valu[i]<cap :
            df.loc[i, 'Extra_Qty'] = 0
            df.loc[i, 'Rem_Cap'] = cap-valu[i]
            
        else :
            df.loc[i, 'Extra_Qty'] = 0
            df.loc[i, 'Rem_Cap'] = 0
            
    df.to_csv("C:/Users/Admin/Desktop/AIS Assignment/Project/Task1.csv", index=False)
    return df

    
```

```python
Task1(1200,dict1)
```

```python
Task1(800,dict1)
```
