```python
#Create a user defined function to accept the demand(as a dictionary) and capacity of the production 
#and returns a dataframe which shows the data with Extra quantity and remaining capacity
#The function should also return the adjusted quantity
```

```python
dict1={'jan': 1200, 'feb': 800, 'mar': 700, 'apr': 900, 'may': 1500, 'jun': 500, 'jul': 800, 'aug': 1400, 'sep': 110, 'oct': 100, 'nov': 85, 'dec': 1150}
dict1
```

```python
def Task1(capacity,demand):
    d=demand
    key=list(d.keys())
    valu=list(d.values())
    cap=capacity
    Month=[]
    Cap=[]
    Dem=[]
    Extra_Qty=[]
    Rem_Cap=[]
    for i in range(len(key)):
        Month.append(key[i])
        Cap.append(cap)
        Dem.append(valu[i])
        
        if valu[i]>cap :
            Extra_Qty.append(valu[i]-cap)
            Rem_Cap.append(0)
            
        elif valu[i]<cap :
            Extra_Qty.append(0)
            Rem_Cap.append(cap-valu[i])
            
        else :
            Extra_Qty.append(0)
            Rem_Cap.append(0)
            
    import pandas as pd
    Table = {'Month':Month,'Dem':Dem,'Cap':Cap,'Extra_Qty':Extra_Qty,'Rem_Cap':Rem_Cap}
    df=pd.DataFrame(Table,index=range(1,13))
    return df
```

```python
Task1(1000,dict1)
```

```python
Task1(1100,dict1)
```
